```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
from keras.layers import InputLayer, Dense, Flatten
from keras.layers import AveragePooling2D, MaxPooling2D
from keras.layers.convolutional import Conv2D
from keras.models import Model, Sequential
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras.preprocessing import image
```

```python
#Replace max pool with average pool as avg pool retains more information
def maxpool_to_avgpool(shape):
    vgg = VGG16(input_shape=shape, weights='imagenet', include_top=False)
    new_model = Sequential()
    new_model.add(InputLayer(input_shape=shape))
    for layer in vgg.layers:
         if(layer.__class__ == MaxPooling2D):
            new_model.add(AveragePooling2D())
         else:
            new_model.add(layer)
    return new_model
```

```python
def vgg_cut(shape,l):
    new_model = maxpool_to_avgpool(shape)
    if l<1 or l>13:
        print("Enter valid numbe of layers")
    model = Sequential()
    n=0
    for layer in new_model.layers:
        model.add(layer)
        if(layer.__class__ == Conv2D):
            n+=1
        if n>=l:
            break
    return model
```

```python
from PIL import Image
path='C:\Users\KIIT\Desktop\Projects\Style_transfer\Target.jpg'
size=(300,300) #Resize image to this size after loading
```

```python
img = Image.open(path)
```

```python
img=img.resize(size)
```

```python
x=np.asarray(img)
```

```python
x.shape
```

```python
#We have to expand dimension as model takes input in form of batches
x = np.expand_dims(x, axis=0)
x.shape
```

```python
x = preprocess_input(x) #Keras function to preprocess input according to VGG16
```

```python
batch_shape = x.shape #Shape of batch tensor (batch_size x H x W x Channels)
shape = x.shape[1:] #shape of image
```

```python
shape
```

```python
plt.imshow(img) #This is the target image
plt.show()
```

```python
model = vgg_cut(shape, 11)
```

```python
for layer in model.layers:
    layer.trainable=False
```

```python
import keras.backend as K #Tensorflow and theano wrapper
```

```python
target = K.variable(model.predict(x))
```

```python
loss = K.mean(K.square(target-model.output))
```

```python
grads = K.gradients(loss, model.input)
```

```python
get_loss_grads = K.function(inputs=[model.input], outputs=[loss]+grads)
```

```python
#bfgs expects input in form of vectors while our get_loss_grad takes 4-D tensors
def loss_and_grads_wrapper(x_vec):
    l,g = get_loss_grads([x_vec.reshape(batch_shape)])
    return l.astype(np.float64), g.flatten().astype(np.float64)
```

```python
loses=[]
```

```python
x = np.random.randn(np.prod(batch_shape)) #Initialize a random image
```

```python
from scipy.optimize import fmin_l_bfgs_b
```

```python
for i in range(20):
    x,l,_ = fmin_l_bfgs_b(func=loss_and_grads_wrapper, x0=x, maxfun=20)
    x=np.clip(x,-127,127)
    print("iter=%s loss=%s" % (i,l))
    loses.append(l)
```

```python
plt.plot(loses)
plt.show()
```
